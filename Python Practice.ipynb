### (1) Write a function that inputs a number and prints the multiplication table of that number

```python
def multiplication_table(number):
# User Input - 
#(1)Number for which multiplication table is required. 
#(2)The range till which the multiplication table should be printed

    if number < 0:
        return ('Inputs are invalid. Please enter non-negative numbers')
    mult_range = int(input('Please enter the range till which the multiplication table will print: '))
    print(f'Multiplication Table for {number} till {mult_range}')
    print('--------------------------------------------------------')
    for i in range(1,mult_range+1):
        print(f'{number}* {i} = ', number * i)

```

```python
# Printing the multiplication table for a sample number -->        
multiplication_table(5)
```

### (2) Write a Program to print twin primes less than 1000

```python
'''

1. Consider element "i" from the range(2,1000)

2. Check element "i" is prime or not

3. If the element "i" is prime, then check element "i+2" is prime or not.

4. If "i+2" element is prime, then print (i, i+2).

'''

for num in range(2,1000):
    if num>2:
        num_twin = num+2
        for i in range(2,num):
            if num%i== 0:
                break
            elif (num_twin) % i == 0:
                break
        else:
            print(num,num_twin)        
```

### (4) Write a program to implement the formulae of permutations and combinations.

```python
def permutations(n,r):
    n_factorial = 1
    denom = n-r
    denom_factorial = 1
    for i in range (1,n+1):
        n_factorial = n_factorial*i
    for j in range(1,denom+1):
        denom_factorial = denom_factorial * j
        permut = n_factorial/denom_factorial
    return print('The possible number of permutations with {} and {} are:'.format(n,r),int(permut))

def combinations(n,r):    
    r_factorial = 1
    for k in range(1,r+1):
        r_factorial = r_factorial * k
        n_factorial = 1
        denom = n-r
        denom_factorial = 1
    for i in range (1,n+1):
        n_factorial = n_factorial*i
    for j in range(1,denom+1):
        denom_factorial = denom_factorial * j
    permut = n_factorial/denom_factorial
    combin = permut/r_factorial
    return print('The possible number of combinations with {} and {} are:'.format(n,r),int(combin))


def perm_comb(n,r):
    if r > n :
        print("Please enter a value of r smaller than n")
    elif r<0 or n<0: 
        print('Enter Non-Negative numbers')
    permutations(n,r)
    combinations(n,r)    
```

### (5) Write a function that converts a decimal number to a binary number

```python
def decimal_to_binary():
    usr_input = input('Enter a number: ')
    num = int(usr_input)
    binary_num = []
    quotient = num
    while num!=0:
        rem = quotient%2
        quotient = num//2
        num = quotient
        binary_num.append(rem)
    binary_num.append(0)
    final_list = binary_num[::-1]
    string_output = ''.join(map(str,final_list))
    return print('\nThe number %s in binary is written as :'%(usr_input),string_output)
```

### (6) Write a function cubesum() that accepts an integer and returns the sum of the cubes of individual digits of that number. Use this function to make functions PrintArmstrong() and isArmstrong() to print Armstrong numbers and to find whether is an Armstrong number

```python
"""
To run the program we need to call the function isArmstrong and pass a number to check 
if the number is an Armstrong number or not.
"""
def cubesum(number):    
# The typecasting Logic used to convert the number into string such that the digits can be separated and stored in a list 
# is taken from link: https://stackoverflow.com/questions/21270320/turn-a-single-number-into-single-digits-python
    num_split = [int(i) for i in str(number)] # Typecasting into string to iterate over it and finally saving into list
    cube_sum = 0
    for num in num_split:
        num_cube = num ** 3
        cube_sum += num_cube
    return cube_sum

def PrintArmstrong(number):
    if number == 1:
        print('Number is an Armstrong number')
    elif number == 2:
        print('Number is not an Armstrong number')

def isArmstrong(number):
    check_Armstrong = cubesum(number)
    if number == check_Armstrong:
        return PrintArmstrong(1)
    else:
        return PrintArmstrong(2)
```

### (7) Write a function prodDigits() that inputs a number and returns the product of digits of that number

```python
def prodDigits(number):
    digit = [int(i) for i in str(number)]
    digit_product = 1
    for j in range(0, len(digit)):
        digit_product *= digit[j]
    print('The product of digits in the number {} is : '.format(number),digit_product)
    return digit_product
```

### (8) Using the function prodDigits() of previous exercise write functions MDR() and MPersistence() that input a number and return its multiplicative digital root and multiplicative persistence respectively

```python
def MDR(number):
    num = number
    if num in range(0,10):
        return print(' The Multiplicative Digital Root of number is : ', num)
    while num>9:
        num = prodDigits(num)
    print("\033[1m"'\nThe Multiplicative Digital Root of number {} is : '.format(number), num)
    return num
```

```python
MDR(86)
```

```python
def MPersistence(number):
    num = number
    counter = 0
    if num in range(0,10):
        return print(' The Multiplicative persistence of number is : ',counter)
    while num>9:
        num = prodDigits(num)
        counter += 1
    print("\033[1m"'\nThe Multiplicative Persistence of number {} is : '.format(number), counter)
    return counter
```

### (9) Write a function sumPdivisors() that finds the sum of proper divisors of a number

```python
def sumPdivisors(num):
# Finding all the factors of a number
    prop_divisor = []
    divisor_sum = 0
    for i in range(1,num): 
        if num%i == 0:
            prop_divisor.append(i)
            divisor_sum = sum(prop_divisor)
    return print('The proper divisors are {} and their sum is {}'.format(prop_divisor,divisor_sum))
```

### (10) Write a program to print all the perfect numbers in a given range

```python
def sumPrdivisors(num):
# Finding all the factors of a number
    pr_divisor = []
    div_sum = 0
    for i in range(1,num): 
        if num%i == 0:
            pr_divisor.append(i)
            div_sum = sum(pr_divisor)
    return div_sum 

def PerfectNumber():
    lower_limit = int(input('Enter the starting number of the range : '))
    upper_limit = int(input('Enter the ending number of the range : '))
    perf_num = []
    
    for i in range(lower_limit, upper_limit+1):
        x = sumPrdivisors(i)
        if x == i:
            perf_num.append(i)
    return print("\033[1m"'\nThe perfect numbers in the range provided are : ',perf_num)
```

### (12) Write a program which can filter odd numbers in a list by using filter function

```python
"""
First we create a function which returns False if the number is an odd number as we want to filter or remove
odd numbers. The filter function takes two arguments - One is a function and the other is a iterable object
"""
usr_list = []
def odd_num(num):
    if num%2 !=0:
        return False
    else:
        return True
# creating a function which takes the user list as input and then filters through the list
def odd_filter():
    count = int(input('Enter the total number of values you want in the list: '))
    while count!=0:
        number = int(input('Enter a number: '))
        count -= 1
        usr_list.append(number)
    for i in usr_list:
        final = list(filter(odd_num,usr_list))
    return print('\nThe filtered list after the removal of odd numbers is: ',final)
```

### (13) Write a program which can map() to make a list whose elements are cube of elements in a given list

```python
## Using the inbuilt map() function in python -->
def element_cube():
    user_list = []
    count = int(input('Enter the total number of values you want in the list: '))
    while count!=0:
        number = int(input('Enter a number: '))
        count -= 1
        user_list.append(number)
        num_cube = list(map(lambda x: x**3, user_list))
    return print('\nThe list of cube of input numbers is: ', num_cube)
```
